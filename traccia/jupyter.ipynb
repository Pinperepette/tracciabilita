```python
import sqlite3

# Connessione al database
def get_db_connection():
    conn = sqlite3.connect('database/tracabilita.db')
    conn.row_factory = sqlite3.Row
    return conn

# Funzione per stampare la struttura e i dati di tutte le tabelle
def print_database_structure_and_data():
    conn = get_db_connection()
    cursor = conn.cursor()

    # Ottenere tutte le tabelle nel database
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()

    # Iterare attraverso le tabelle e stampare la struttura e i dati
    for table in tables:
        table_name = table['name']
        print(f"\n--- Struttura e dati della tabella '{table_name}' ---\n")

        # Stampare la struttura della tabella
        cursor.execute(f"PRAGMA table_info({table_name});")
        columns = cursor.fetchall()
        print("Colonne:")
        for col in columns:
            print(f" - {col['name']} (Tipo: {col['type']})")

        # Stampare i dati della tabella
        cursor.execute(f"SELECT * FROM {table_name};")
        rows = cursor.fetchall()
        print("\nDati:")
        if rows:
            for row in rows:
                print(dict(row))
        else:
            print("Nessun dato presente in questa tabella.")

    conn.close()

# Esegui lo script

print_database_structure_and_data()

```
